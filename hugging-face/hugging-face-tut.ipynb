```python
from huggingface_hub import login,whoami
import os

hugging_face_api_key = os.getenv("HUGGINGFACE_API_KEY")

if hugging_face_api_key:
    login(token=hugging_face_api_key)
else:
    print("HUGGINGFACE_API_KEY not found in environment variables.")
```

### whoami

```python
user = whoami(token=hugging_face_api_key)
print(user["name"])
```

### Use Hugging face Models locally

```python
import torch
from diffusers import DiffusionPipeline
import imageio

# Load model
pipe = DiffusionPipeline.from_pretrained(
    "SulphurAI/Sulphur-2-base",
    torch_dtype=torch.float16
)

# Move to GPU
pipe = pipe.to("cuda")

# Prompt
prompt = "A futuristic cyberpunk city at night with flying cars"

# Generate video frames
result = pipe(
    prompt=prompt,
    num_inference_steps=30
)

frames = result.frames[0]

# Save video
imageio.mimsave(
    "output.mp4",
    frames,
    fps=8
)

print("Video saved!")
```
