```python
from langchain_tavily import TavilySearch
from langgraph.graph import StateGraph,START,END,MessagesState
from langgraph.graph.message import add_messages
from typing import TypedDict
from typing_extensions import Annotated
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv

from langgraph.checkpoint.memory import MemorySaver
from langgraph.prebuilt import ToolNode
from langchain.tools import tool

# loading dotenv
load_dotenv()
```

```python
# state definition
class AgentState(MessagesState):
    next_agent:str   # which agent should go next
```

```python
# Tool Creation

@tool
def websearch_tool(query:str) -> str:
    """ Search the web Information """

    search_data = TavilySearch(
        max_results=3,
    )
    results = search_data.invoke(query)

    return str(results)

@tool
def create_summary(content:str) -> str:
    """ Writes the summary of the content """
    summary = f"Summary of findings:\n\n{content[:500]}..."

    return summary
```
