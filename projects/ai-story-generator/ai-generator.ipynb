## AI Story Generator ✨

```python
from typing import Annotated
from typing_extensions import TypedDict
from langgraph.graph import START,END,StateGraph
from dotenv import load_dotenv
from langgraph.graph.message import add_messages
import os
load_dotenv()
```

```python
## LLM Initialization
from langchain_groq import ChatGroq

llm_model = ChatGroq(
    model="openai/gpt-oss-20b"
)
```

```python
# Creating an Message StateGraph
class StoryState(TypedDict):
    topic: str
    story: str
    improved_story: str
    title: str
```

```python
# first Node

def story_generator(state:StoryState):
    topic = state["topic"]

    ai_response = llm_model.invoke(
         f"Write a short creative story about {topic}"
    )

    return {
        "story":ai_response.content
    }
```

```python
# 2nd Node
def improve_story(state:StoryState):

    story = state["story"]

    ai_response = llm_model.invoke(
        f"improve the story and make it cinematic:\n {story}"
    )

    return {
        "improved_story":ai_response.content
    }
```

```python
def generate_title(state:StoryState):

    story = state["improved_story"]

    ai_response = llm_model.invoke(
        f"generate a title:\n {story}"
    )
    return {
        "title":ai_response.content
    }
```

```python
# build a graph

story_graph = StateGraph(StoryState)
```

```python
story_graph.add_node("generate_story",story_generator)
story_graph.add_node("improve_story",improve_story)
story_graph.add_node("generate_title",generate_title)
```

```python
story_graph.add_edge(START, "generate_story")

story_graph.add_edge(
    "generate_story",
    "improve_story"
)

story_graph.add_edge(
    "improve_story",
    "generate_title"
)

story_graph.add_edge(
    "generate_title",
    END
)
```

```python
ai_story_agent =story_graph.compile()
```

```python
result = ai_story_agent.invoke({
    "topic": "Space Pirates"
})
```
